14. Read all json files with the video categories

```python
import glob
import json
import os

folder_path = "file_category"   # cartella con i json
json_files = glob.glob(os.path.join(folder_path, "*_category_id.json"))

categories_dict = {}

for file in json_files:
    # Estrae il country code dal nome file (es. CA_category_id.json → CA)
    country = os.path.basename(file).split("_")[0]

    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)

    # Prendiamo SOLO le categorie assignable
    assignable_ids = {
        int(item["id"])
        for item in data["items"]
        if item["snippet"]["assignable"] is True
    }

    categories_dict[country] = assignable_ids
```

```python
categories_dict
```

15. For each country, determine how many videos have a category that is not assignable.

```python
# Controllo se la categoria del video è assegnabile nel paese
youtubedf['assignable'] = youtubedf.apply(
    lambda row: row['category_id'] in categories_dict.get(row['country'], set()),
    axis=1
)

# Video non assegnabili
non_assignable_videos = youtubedf[~youtubedf['assignable']]
```

```python
non_assignable_videos['assignable']
```

```python
# Conteggio per paese
count_non_assignable = (
    non_assignable_videos
    .groupby('country')
    .size()
    .reset_index(name='num_non_assignable_videos')
)

count_non_assignable
```
